# Data preparation

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load the iris dataset
iris = load_iris()
X = iris['data']
y = iris['target']
names = iris['target_names']
feature_names = iris['feature_names']

# Scale data to have mean 0 and variance 1
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data set into training and testing
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=2)
```

# Visualising dataset

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
for target, target_name in enumerate(names):
    X_plot = X[y == target]
    ax1.plot(X_plot[:, 0], X_plot[:, 1],
             linestyle='none',
             marker='o',
             label=target_name)
ax1.set_xlabel(feature_names[0])
ax1.set_ylabel(feature_names[1])
ax1.axis('equal')
ax1.legend();

for target, target_name in enumerate(names):
    X_plot = X[y == target]
    ax2.plot(X_plot[:, 2], X_plot[:, 3],
             linestyle='none',
             marker='o',
             label=target_name)
ax2.set_xlabel(feature_names[2])
ax2.set_ylabel(feature_names[3])
ax2.axis('equal')
ax2.legend();
```

# MLP for multiclass classification

```python
import torch
import torch.nn.functional as F
import torch.nn as nn
from torch.autograd import Variable
import tqdm
```

```python
# Defining the model architecture
class Model(nn.Module):
    def __init__(self, input_dim):
        super(Model, self).__init__()
        self.layer1 = nn.Linear(input_dim, 50)
        self.layer2 = nn.Linear(50, 50)
        self.layer3 = nn.Linear(50, 3)

    def forward(self, x):
        x = F.relu(self.layer1(x))
        x = F.relu(self.layer2(x))
        x = F.softmax(self.layer3(x), dim=1)
        return x
```

```python
# Instantiating the model, using Adam optimiser, and Cross Entropy Loss, which is quite commonlu used for classification tasks.
model     = Model(X_train.shape[1])
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss_fn   = nn.CrossEntropyLoss()
```

# Train the model

```python
# Train for 100 epochs
EPOCHS  = 100
X_train = torch.from_numpy(X_train).float()
X_test= torch.from_numpy(X_test).float()
y_test = torch.from_numpy(y_test)
y_train = torch.from_numpy(y_train)


loss_list     = np.zeros((EPOCHS,))
accuracy_list = np.zeros((EPOCHS,))

for epoch in tqdm.trange(EPOCHS):
    y_pred = model(X_train)
    loss = loss_fn(y_pred, y_train)
    loss_list[epoch] = loss.item()

    # Zero gradients
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    with torch.no_grad():
        y_pred = model(X_test)
        correct = (torch.argmax(y_pred, dim=1) == y_test).type(torch.FloatTensor)
        accuracy_list[epoch] = correct.mean()
```

# Plot training progress

```python
fig, (ax1, ax2) = plt.subplots(2, figsize=(12, 6), sharex=True)

ax1.plot(accuracy_list)
ax1.set_ylabel("validation accuracy")
ax2.plot(loss_list)
ax2.set_ylabel("validation loss")
ax2.set_xlabel("epochs");
```

It sounds like you’re experimenting with a neural network and trying to understand how changes to its architecture and parameters affect its performance. Here's an explanation of what each modification might do and how to observe changes:

1. Changing the Number of Hidden Layers:

Effect: Adding more hidden layers increases the network's capacity to model complex functions but might lead to overfitting if the dataset is small. Fewer layers may simplify the model, reducing overfitting but potentially underfitting the data.

Observation: Look at the training and validation accuracy/loss. If you see a significant gap (high training accuracy but low validation accuracy), it might indicate overfitting with too many layers.



2. Changing the Number of Neurons in the Hidden Layer:

Effect: Increasing the number of neurons gives the layer more capacity to capture patterns, while fewer neurons reduce model complexity. Too many neurons can lead to overfitting, while too few can result in underfitting.

Observation: Check how the loss and accuracy behave during training. A well-sized layer should lead to a balance between good training and validation performance.



3. Using a Different Activation Function:

Effect: The activation function determines the non-linear transformations applied in the network. For example, switching from ReLU to sigmoid or tanh can affect convergence and gradient behavior.

Observation: Compare convergence speed and the final accuracy or loss. Some functions (e.g., ReLU) are faster to train but might lead to issues like dead neurons, while others (e.g., sigmoid) might suffer from vanishing gradients.




What to Observe:

Training Metrics: Accuracy or loss during training (does it plateau, diverge, or converge faster?).

Validation Metrics: Check for overfitting or underfitting by comparing performance on unseen data.

Learning Curve: Visualize the training and validation loss/accuracy over epochs to identify patterns.
